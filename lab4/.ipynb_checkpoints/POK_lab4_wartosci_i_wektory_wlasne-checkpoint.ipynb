## Zadanie 2. Wartości i wektory własne macierzy
\begin{bmatrix}
3 &1  &2 \\ 
1 &8  &1 \\ 
2 &1  &1
\end{bmatrix}

Dla powyższej macierzy wyznaczyć wartości własne macierzy i wektory własne macierzy z wykorzystaniem dowolnej metody.

1. Macierz kwadratowa

$A = \begin{bmatrix}
3 &1  &2 \\ 
1 &8  &1 \\ 
2 &1  &1 
\end{bmatrix}$


2. Równanie z macierzą

$A_{\lambda} = \begin{bmatrix}
3 &1  &2 \\ 
1 &8  &1 \\ 
2 &1  &1 
\end{bmatrix} - \lambda * \begin{bmatrix}
1 &0  &0 \\ 
0 &1  &0 \\ 
0 &0  &1 
\end{bmatrix} = \begin{bmatrix}
3 &1  &2 \\ 
1 &8  &1 \\ 
2 &1  &1 
\end{bmatrix} - \begin{bmatrix}
\lambda &0  &0 \\ 
0 &\lambda  &0 \\ 
0 &0  &\lambda 
\end{bmatrix}$

$A_{\lambda}= \begin{bmatrix}
3-\lambda &1  &2 \\ 
1 &8-\lambda  &1 \\ 
2 &1  &1-\lambda 
\end{bmatrix}$


3. Wyznacznik macierzy

$\begin{vmatrix}
3-\lambda &1  &2 \\ 
1 &8-\lambda  &1 \\ 
2 &1  &1-\lambda 
\end{vmatrix} = -8 - 29 \lambda + 12 \lambda^2 - \lambda^3$


4. Policzenie pierwiastków wyznacznika macierzy

$0 = -8 - 29 \lambda + 12 \lambda^2 - \lambda^3$


```python
### POLICZENIE POWYŻSZEGO RÓWNANIA ###

import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

# zdefiniowe współczynników
coef = [-8, -29, 12, -1]
coef2= [-1, 12, -29, -8]

# zdefiniowanie funkcji
F=poly.Polynomial(coef) # to było pisane w Stanach, dlatego na odwrót

# odnalezienie pierwiastków
roots = np.roots(coef2) # to było pisane w Europie, dlatego normalny zapis
print(f"Pierwiastki funkcji:\n{F} \nwynoszą:\n{roots}")

# wizualizacja funkcji 
x=np.arange(-2, 10, 0.01)
y=F(x)
plt.figure(figsize = (15,6))
plt.plot(x, y, label = "Funkcja wyznacznika macierzy") 
plt.plot(roots[0], F(roots[0]), 'r*', label = "Pierwszy pierwiastek") 
plt.plot(roots[1], F(roots[1]), 'k*', label = "Drugi pierwiastek")
plt.plot(roots[2], F(roots[2]), 'y*', label = "Trzeci pierwiastek")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title("Wartości własne macierzy")
plt.show()
```

***Powtarzając procedurę dla kolejnych wartości $\lambda$ można uzyskać wektory własne macierzy wynoszące:***
- $v_1\approx[1.26808, 4.92443, 1]$
- $v_2\approx[1.71711, -0.645237, 1]$
- $v_3\approx[-0.600569, -0.0484185, 1]$

```python
### SPRAWDZENIE
# wprowadzenie macierzy
A = np.array([[3, 1, 2], [1, 8, 1], [2, 1, 1]])
print("Macierz:")
print(A)

# policzenie wartości własnych oraz wektorów
w,v = np.linalg.eig(A)

# przeskalowanie wektorów do wyników podobnych do metody analitycznej
v[:, 0] = np.multiply(v[:, 0], 1/v[2, 0])
v[:, 1] = np.multiply(v[:, 1], 1/v[2, 1])
v[:, 2] = np.multiply(v[:, 2], 1/v[2, 2])

# wyświetlanie wyników
print(f"\nWartości własne macierzy:\n{w}")
print(f"\nWektory własne macierzy wynoszą:\n{v}")
```
