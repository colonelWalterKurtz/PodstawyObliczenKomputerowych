Dla powyższych danych odnaleźć wartość wielomianu w punkcie **x=21** z wykorzystaniem dwóch metod:
- Interpolacja wielomianowa
- Interpolacja za pomocą funkcji sklejanych

### - Interpolacja wielomianowa (Lagrange'a)
- $P_1(x)=\frac{(x-x_2)(x-x_3)(x-x_4)(x-x_5)}{(x_1-x_2)(x_1-x_3)(x_1-x_4)(x_1-x_5)}=\frac{(x-20)(x-30)(x-40)(x-50)}{(10-20)(10-30)(10-40)(10-50)}=\frac{x^4}{240000}-\frac{7x^3}{12000}+\frac{71x^2}{2400}-\frac{77x}{120}+5$


- $P_2(x)=\frac{(x-x_1)(x-x_3)(x-x_4)(x-x_5)}{(x_2-x_1)(x_2-x_3)(x_2-x_4)(x_2-x_5)}=\frac{(x-10)(x-30)(x-40)(x-50)}{(20-10)(20-30)(20-40)(20-50)}=\frac{x^4}{60000}-\frac{13x^3}{6000}+\frac{59x^2}{600}-\frac{107x}{60}+100$


- $P_3(x)=\frac{(x-x_1)(x-x_2)(x-x_4)(x-x_5)}{(x_3-x_1)(x_3-x_2)(x_3-x_4)(x_3-x_5)}=\frac{(x-10)(x-20)(x-40)(x-50)}{(30-10)(30-20)(30-40)(30-50)}=\frac{x^4}{40000}-\frac{3x^3}{1000}+\frac{49x^2}{400}-\frac{39x}{20}+10$


- $P_4(x)=\frac{(x-x_1)(x-x_2)(x-x_3)(x-x_5)}{(x_4-x_1)(x_4-x_2)(x_4-x_3)(x_4-x_5)}=\frac{(x-10)(x-20)(x-30)(x-50)}{(40-10)(40-20)(40-30)(40-50)}=-\frac{x^4}{60000}+\frac{11x^3}{6000}-\frac{41x^2}{600}+\frac{61x}{60}-5$


- $P_5(x)=\frac{(x-x_1)(x-x_2)(x-x_3)(x-x_4)}{(x_5-x_1)(x_5-x_2)(x_5-x_3)(x_5-x_4)}=\frac{(x-10)(x-20)(x-30)(x-40)}{(50-10)(50-20)(50-30)(50-40)}=\frac{x^4}{240000}-\frac{x^3}{2400}+\frac{7x^2}{480}-\frac{5x}{24}+1$

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt

# znane punkty
x = [10, 20, 30, 40, 50]
y = [0.98, 0.93, 0.86, 0.76, 0.64]

# potrzeba utworzyć 5 funkcji skoro pięć punktów
#P1_coeff = [1/240000*1200000, -1/240000*154000, 1/240000*7100, -1/240000*140, 1/240000]
P1_coeff = [5, -77/120, 71/2400, -7/12000, 1/240000]
P2_coeff = [-10, 107/60, -59/600, 13/6000, -1/60000]
P3_coeff = [10, -39/20, 49/400, -3/1000, 1/40000]
P4_coeff = [-5, 61/60, -41/600, 11/6000, -1/60000]
P5_coeff = [1, -5/24, 7/480, -1/2400, 1/240000]


# get the polynomial function
P1 = poly.Polynomial(P1_coeff)
P2 = poly.Polynomial(P2_coeff)
P3 = poly.Polynomial(P3_coeff)
P4 = poly.Polynomial(P4_coeff)
P5 = poly.Polynomial(P5_coeff)
x_new = np.arange(0, 60, 0.1)

fig = plt.figure(figsize = (15, 6))
plt.plot(x_new, P1(x_new), 'b', label = "P1")
plt.plot(x_new, P2(x_new), 'r', label = "P2")
plt.plot(x_new, P3(x_new), 'g', label = "P3")
plt.plot(x_new, P4(x_new), 'c', label = "P4")
plt.plot(x_new, P5(x_new), 'y', label = "P5")

plt.plot(x, np.ones(len(x)), 'ko', x, np.zeros(len(x)), 'ko')
plt.title("Wizualizacja funkcji składowych interpolacji")
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.axis([0, 60, -10, 10])
plt.legend()
plt.show()
```

```python
L = 0.98*P1 + 0.93*P2 + 0.86*P3 + 0.76*P4 + 0.64*P5
interpolate = L(21)
print(f"Zinterpolowana wartość punktu x=21 wynosi y={interpolate}")
# 21 stopni w radianach to 7pi/60
wlasc_wart=np.cos(7*np.pi/60)
blad_wzgledny=(wlasc_wart-interpolate)/wlasc_wart*100
print(f"Właściwa wartość wynosi {wlasc_wart}, a więc błąd względny interpolacji wynosi {blad_wzgledny}%")

fig = plt.figure(figsize = (15, 6))
plt.plot(x_new, L(x_new), 'b', x, y, 'ro')
plt.plot(21, L(21), 'k*', label = "Interpolowana wartość punktu")
plt.plot(21, np.cos(7*np.pi/60), 'c*', label="Właściwa wartość punktu")
plt.title("Wizualizacja interpolacji wielomianowej Lagrange")
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
### SPRAWDZENIE ###
from scipy.interpolate import lagrange

f = lagrange(x, y)
print(f"Wartość zinterpolowana za pomocą wbudowanej funkcji intepolacji {f(np.cos(7*np.pi/60))}")
interpolate_scipy=f(21)
blad_wzgledny2 = np.abs((wlasc_wart-interpolate_scipy)/wlasc_wart*100)
print(f"Właściwa wartość wynosi {wlasc_wart}, a więc błąd względny interpolacji wynosi {blad_wzgledny2}%")

print(f"Różnica pomiędzy wartością obliczoną za pomocą autorskiej implementacji, a wbudowanej w pakiet scipy funkcji lagrange wynosi: {interpolate-interpolate_scipy}")
```

### - Interpolacja funkcji sklejanych

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

# znane punkty
x = [10, 20, 30, 40, 50]
y = [0.98, 0.93, 0.86, 0.76, 0.64]

# use bc_type = 'natural' adds the constraints as we described above
CS = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 60, 100)
y_new = CS(x_new)

# szukany punkt
punkt_x=21
punkt_y=CS(21)
print(f"Szukana wartość w położeniu x=21, zinterplowana za pomoca funkcji sklejanych wynosi y={punkt_y}")
wlasc_wart=np.cos(7*np.pi/60)
blad_wzgledny3 = np.abs((wlasc_wart-punkt_y)/wlasc_wart*100)
print(f"Błąd względny pomiędzy wartością rzeczywistą({wlasc_wart}) wynosi: {blad_wzgledny3}%")

plt.figure(figsize = (15,6))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.plot(21, CS(21), 'k*', label = "Interpolowana wartość punktu")
plt.plot(21, np.cos(7*np.pi/60), 'c*', label="Właściwa wartość punktu")
plt.plot()
plt.title('Interpolacja funkcji sklejanych')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
